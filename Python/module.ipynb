# date and time : A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.

```python
# A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.
import datetime
a = datetime.datetime.now()
print(a)
print(a.year , a.month , a.day , a.hour , a.minute , a.second , a.microsecond)

print(a.strftime("%A"))

#dir(datetime)
b = datetime.datetime.now()

# Weekday, full version
print(b.strftime("%A"))
# Weekday, short version
print(b.strftime("%a"))
# Month name, short version
print(b.strftime("%b"))
# Month name, full version
print(b.strftime("%B"))
# Local version of date and time
print(b.strftime("%c"))
print(b.strftime("%C"))
# Day of month 01-31
print(b.strftime("%D"))
print(b.strftime("%d"))

print(b.strftime("%e"))
print(b.strftime("%F"))
# Microsecond 000000-999999
print(b.strftime("%f"))
print(b.strftime("%G"))
print(b.strftime("%g"))
# Hour 00-23
print(b.strftime("%H"))
print(b.strftime("%h"))
# Hour 00-12
print(b.strftime("%I"))

# Day number of year 001-366
print(b.strftime("%j"))

print(b.strftime("%k"))

print(b.strftime("%l"))
# Minutes 00-59
print(b.strftime("%M"))
# Month as a number 01-12
print(b.strftime("%m"))

print(b.strftime("%n"))


# AM/PM
print(b.strftime("%P"))
print(b.strftime("%p"))


print(b.strftime("%R"))
print(b.strftime("%r"))
# Second 00-59
print(b.strftime("%S"))
print(b.strftime("%s"))
print(b.strftime("%t"))
print(b.strftime("%T"))
# Week number of year, Sunday as the first day of week, 00-53
print(b.strftime("%U"))
print(b.strftime("%u"))
print(b.strftime("%V"))

# Week number of year, Monday as the first day of week, 00-53
print(b.strftime("%W"))
# Weekday as a number 0-6, 0 is Sunday
print(b.strftime("%w"))
# Local version of time
print(b.strftime("%X"))
# Local version of date
print(b.strftime("%x"))
# Year, full version
print(b.strftime("%Y"))
# Year, short version, without century
print(b.strftime("%y"))
# Timezone
print(b.strftime("%Z"))
#  UTC offset
print(b.strftime("%z"))
# A % character
print(b.strftime("%%"))

```

# Python | __import__() function
    While writing a code, there might be a need for some specific modules. So we import those modules by using a single line code in Python.

    But what if the name of the module needed is known to us only during runtime? How can we import that module? One can use the Pythonâ€™s inbuilt __import__() function. It helps to import modules in runtime also.
    
# Syntax: __import__(name, globals, locals, fromlist, level)

    Parameters:
    name : Name of the module to be imported
    globals and locals : Interpret names
    formlist : Objects or submodules to be imported (as a list)
    level : Specifies whether to use absolute or relative imports. Default is -1(absolute and relative).

```python
pip install numpy
```

```python
np = __import__('numpy', globals(), locals(), [], 0) 
  
# array from numpy 
a = np.array([1, 2, 3]) 
  
# prints the type 
print(type(a)) 
```

```python
np = __import__('numpy', globals(), locals(), ['complex', 'array'], 0) 
  
comp = np.complex
arr = np.array 
print(type(comp))
print(type(arr))
```

# Python Urllib Module

    Urllib module is the URL handling module for python. It is used to fetch URLs (Uniform Resource Locators). It uses the urlopen function and is able to fetch URLs using a variety of different protocols.

    Urllib is a package that collects several modules for working with URLs, such as:

    urllib.request for opening and reading.
    urllib.parse for parsing URLs
    urllib.error for the exceptions raised
    urllib.robotparser for parsing robot.txt files

```python
import urllib.request
url = urllib.request.urlopen("https://www.geeksforgeeks.org/")
print(url.read())
```

```python
from urllib.parse import * parse_url = urlparse('https://www.geeksforgeeks.org / python-langtons-ant/') 
print(parse_url) 
print("\n") 
unparse_url = urlunparse(parse_url) 
print(unparse_url) 
```

    urllib.parse.urlparse	Separates different components of URL
    urllib.parse.urlunparse	Join different components of URL
    urllib.parse.urlsplit	It is similar to urlparse() but doesnâ€™t split the params
    urllib.parse.urlunsplit	Combines the tuple element returned by urlsplit() to form URL
    urllib.parse.urldeflag	If URL contains fragment, then it returns a URL removing the fragment.

# urllib.robotparser

    This module contains a single class, RobotFileParser. This class answers question about whether or not a particular user can fetch a URL that published robot.txt files. Robots.txt is a text file webmasters create to instruct web robots how to crawl pages on their website. The robot.txt file tells the web scraper about what parts of the server should not be accessed.

```python
import urllib.robotparser as rb 
  
bot = rb.RobotFileParser() 
  
# checks where the website's robot.txt file reside 
x = bot.set_url('https://www.geeksforgeeks.org / robot.txt') 
print(x) 
  
# reads the files 
y = bot.read() 
print(y) 
  
# we can crawl the main site 
z = bot.can_fetch('*', 'https://www.geeksforgeeks.org/') 
print(z) 
  
# but can not crawl the disallowed url 
w = bot.can_fetch('*', 'https://www.geeksforgeeks.org / wp-admin/') 
print(w) 
```

# pprint : Data pretty printer in Python

    This article is about a pretty useful built-in module in Python, pprint.

    The pprint module provides a capability to â€œpretty-printâ€ arbitrary Python data structures in a well-formatted and more readable way!

```python
pip install requests
```

```python
import requests 
  
def geocode(address): 
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    resp = requests.get(url, params = {'address': address}) 
    return resp.json() 
  
# calling geocode function 
data = geocode('India gate') 
  
# printing json response 
print(data) 
```

```python
import requests 
from pprint import pprint 
  
def geocode(address): 
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    resp = requests.get(url, params = {'address': address}) 
    return resp.json() 
  
# calling geocode function 
data = geocode('India gate') 
  
# pretty-printing json response 
pprint(data) 
```

# Python | Generate QR Code using pyqrcode module

    Letâ€™s see how to generate QR code in Python using pyqrcode module.

    pyqrcode module is a QR code generator. The module automates most of the building process for creating QR codes. This module attempts to follow the QR code standard as closely as possible. The terminology and the encodings used in pyqrcode come directly from the standard.


```python
pip install pyqrcode
```

```python
import pyqrcode 
from pyqrcode import QRCode 
  
  
# String which represent the QR code 
s = "www.geeksforgeeks.org"
  
# Generate QR code 
url = pyqrcode.create(s) 
  
# Create and save the png file naming "myqr.png" 
url.svg("myqr.svg", scale = 8) 
```
